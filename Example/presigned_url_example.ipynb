```python
from aws_s3_resource.s3_object import S3Object
import boto3
from pprint import pprint
```

```python

# Params
```

```python
bucket_name = 'sealr-v2-development'
thumbnail_key_extension = 'jpg'
expiration = 12  # [S]
```

```python
session = boto3.Session(profile_name='default')
dev_client = session.client('s3')

list_buckets = dev_client.list_buckets()
list_objects = dev_client.list_objects(Bucket=bucket_name)
    
pprint(list_objects)
```

```python
small_thumbnails = []
large_thumbnails = []
for key in list_objects['Contents']:
    thumb_kay = key['Key'].split('-')

    if 'small' in thumb_kay:
        small_thumbnails.append(key['Key'])

    if 'large' in thumb_kay:
        large_thumbnails.append(key['Key'])

# pprint(large_thumbnails)
```

```python
# small-thumbnail

if len(small_thumbnails) > 0:
    presigned_url = S3Object.generate_presigned_url(
        bucket_name=bucket_name,
        object_name=small_thumbnails[0],
        expiration=expiration
    )

    print(presigned_url)
```

```python
# large-thumbnail

if len(large_thumbnails) > 0:
    presigned_url = S3Object.generate_presigned_url(
        bucket_name=bucket_name,
        object_name=large_thumbnails[0],
        expiration=expiration
    )

    print(presigned_url)
```
